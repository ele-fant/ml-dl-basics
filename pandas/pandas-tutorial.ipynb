# Pandas

Pandas is a simple to use data analysis tool that is integrated with many other data science and ml Python tools. Pandas helps getting (raw) data ready for ml models.

## Series, DataFrames & CSVs

```python
import pandas as pd
```

### Series

A `Series` is a 1-dimensional column of information. Series can be used to create `DataFrames` (see below). A series takes in a list to be created.

```python
# 2 main data types: 1. Series
series = pd.Series(["BMW", "Toyota", "Mercedes"])
series
```

```python
# Series are 1-dimensional columns
```

```python
colours = pd.Series(["red", "blue", "green"])
```

### DataFrame

A `DataFrame` is a 2-dimensional table of `Series`. A data frame takes a `Dictionary` as input to be created.

`axis = 0` --> refers to row

`axis = 1` --> refers to column

```python
car_df = pd.DataFrame({"make": series, "colour": colours})
car_df
```

### Adding columns

```python
car_sales = pd.read_csv("car-sales.csv")
```

```python
# column from Series
seats = pd.Series([5,5,5,5,5])

car_sales["Seats"] = seats
```

```python
car_sales
```

```python
car_sales["Seats"].fillna(5, inplace=True)
```

```python
car_sales
```

```python
# create column from Python list
# has to be same length (!)
fuel_economy = [5.2, 4.4, 6.7, 3.2, 3.9, 5.2, 4.4, 6.7, 3.2, 3.9]

car_sales["fuel-per-100-km"] = fuel_economy
```

```python
car_sales
```

```python
car_sales["total-fuel-used"] = car_sales["Odometer (KM)"] / 100 * car_sales["fuel-per-100-km"]
```

```python
car_sales
```

```python
# column from scalar
car_sales["wheels"] = 4
car_sales
```

```python
# delete column (using axis=1)
car_sales.drop("wheels", axis=1, inplace=True)
```

```python
car_sales
```

```python
car_sales["wheels"] = 4
car_sales["safety_passed"] = True
```

```python
# rename column
car_sales.rename(columns={"Odometer (KM)": "odometer_km", "fuel-per-100-km": "fuel_per_100_km"}, inplace=True)
```

```python
car_sales
```

```python
car_sales.columns = car_sales.columns.str.lower()
car_sales.columns = car_sales.columns.str.replace('[\-]', '_')
```

```python
car_sales
```
